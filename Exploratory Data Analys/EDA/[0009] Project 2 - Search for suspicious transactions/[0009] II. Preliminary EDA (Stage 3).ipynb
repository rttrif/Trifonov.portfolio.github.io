Таблица Bank connections, набора FinCEN и Transactions map имеют общие предикторы icij_sar_id, filer_org_name_id, filer_org_name. Выполним слияние по данным предикторам. 

```python
# Внутреннее объединение переменных fc_tm и  fc_bc по переменным icij_sar_id, filer_org_name_id, filer_org_name
# Введем переменную fc = FinCEN
fc = pd.merge(fc_tm, fc_bc, on =['icij_sar_id','filer_org_name_id','filer_org_name' ], how='inner')

# Проверка результата 
fc.head().T
```

```python
# Фильтрация набора объеденного набора  FinCEN по дате для создания  сопоставимого периода наблюдений 
# Введем переменную fc_period = FinCEN за переод с 2015 по 2017 годы
fc_period = fc.loc[fc['begin_date']>='2015-01-01']

# Проверка результата 
fc_period.head().T
```

Анализ графика сопоставления по времении begin_date и time с transaction_amount_usd и amount_transactions показывает, что по этим переменным наборы не пересекаются в полной мере, поэтому выполним слияние наборов по времени оставив transaction_amount_usd и amount_transactions без изменений. 

```python
# Переименование  time в begin_date для выполнения внутреннего слияния по данной переменной 
blt = blt.rename({'time': 'begin_date'}, axis='columns')

# Проверка результата 
blt.info()
```

```python
# Слияние набора  FinCEN за переод с 2015 по 2017 годы с набором Bitcoin large transactions по перменной  begin_date
# sar_blt = Match suspicious bank transactions to blockchain transaction hashes
sar_blt = fc_period.merge(blt,  how='inner')

# Проверка результата
sar_blt.head().T
```

```python
# Оценка качества результата 
def verify_data_quality(evaluated_dataset):
    '''
    ФУНКЦИЯ КОМПЛЕКСНОЙ ОЦЕНКИ КАЧЕСТВА ДАННЫХ
    '''
    # Проверка типов данных
    print("Типы данных:\n{}\n".format(evaluated_dataset.dtypes) + '_' * 50+ '\n')
   
    # Проверка наличия значений NaN
    print("Общее количество пропущенных значений:\n{}\n".format(evaluated_dataset.isnull().sum().sum()) + '_' * 50+ '\n')
    
    evaluated_dataset.isnull().sum().plot(kind='bar')
    # Add a title and show the plot.
    plt.title('Распределение пропущенных значений по столбцам')
    # Create tick mark labels on the Y axis and rotate them.
    plt.xticks(rotation = 90)
    # Create X axis label.
    plt.xlabel("Columns")
    # Create Y axis label.
    plt.ylabel("NaN Values");
    
    # Количество значений, отличных от NaN
    print("Количество значений, отличных от NaN:\n{}\n".format(evaluated_dataset.count().sum()) + '_' * 50+ '\n')
    
    # Поиск дублирующих строк 
    print("Дублирующих строк:\n{}\n".format(evaluated_dataset.duplicated().sum()) + '_' * 50+ '\n')
    
    # Форма нарбора данных
    print("Форма нарбора данных:\n{}\n".format(evaluated_dataset.shape) + '_' * 50+ '\n')
   
    # Тип индекса набора данных
    print("Тип индекса набора данных:\n{}\n".format(evaluated_dataset.index) + '_' * 50+ '\n')
```

```python
verify_data_quality(sar_blt)
```

Удалим дублирующие значение и переустановим индекс итогового набора 

```python
# Уделение дублирующих значений и переустановка индекса 
sar_blt = sar_blt.reset_index(drop=True).drop_duplicates()

# Проверка результата 
verify_data_quality(sar_blt)
```

#### Корреляционный анализ

```python
# Постороение диагональной корриляционной матрицы.

sns.set_theme(style="white")

# Compute the correlation matrix
corr = sar_blt.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(15, 15))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(150, 275, s=80, l=55, n=9, )

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0, annot=True, fmt='.2f',
            square=True, linewidths=2, cbar_kws={"shrink": .5})
```
