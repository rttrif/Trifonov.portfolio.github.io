## SECTION 4.2 DATA TRANSFORMATION


````python
delivery_windows = delivery_windows.rename(columns={"id": "delivery_window_id"})
delivery_windows.info()
````

````python
orders_delivery_windows = pd.merge(orders, delivery_windows, on=['delivery_window_id','store_id'], how='left')
orders_delivery_windows.info()
````

````python
stores = stores.rename(columns={"id": "store_id"})
stores.info()
````

````python
orders_delivery_windows_stores =  pd.merge(orders_delivery_windows, stores, on='store_id', how='left')
orders_delivery_windows_stores.info()
````

````python
addresses = addresses.rename(columns={"id": "ship_address_id"})
addresses.info()
````

````python
orders_delivery_windows_stores_addresses =  pd.merge(orders_delivery_windows_stores, addresses, on='ship_address_id', how='left')
orders_delivery_windows_stores_addresses.info()
````

````python
replacements = replacements.rename(columns={"item_id": "item_id_replacements", "order_id":"id", 'state':"state_replacements"})
replacements.info()
````

````python
orders_delivery_windows_stores_addresses_replacements = pd.merge(orders_delivery_windows_stores_addresses, replacements, 
                                                                 on='id', how='left')
orders_delivery_windows_stores_addresses_replacements.info()
````

````python
cancellations = cancellations.rename(columns={"item_id": "item_id_cancellations", "order_id":"id", 'state':"state_cancellations"})
cancellations.info()
````

````python
orders_delivery_windows_stores_addresses_replacements_cancellations = pd.merge(orders_delivery_windows_stores_addresses_replacements, cancellations, 
                                                                on='id', how='left')
orders_delivery_windows_stores_addresses_replacements_cancellations.info()
````

````python
sber_market_db = orders_delivery_windows_stores_addresses_replacements_cancellations.reset_index(drop=True)
sber_market_db.head().T
````

````python
def verify_data_quality(evaluated_dataset):
    '''
    ФУНКЦИЯ КОМПЛЕКСНОЙ ОЦЕНКИ КАЧЕСТВА ДАННЫХ
    '''
    # Проверка типов данных
    print("Типы данных:\n{}\n".format(evaluated_dataset.dtypes) + '_' * 50+ '\n')
   
    # Проверка наличия значений NaN
    print("Общее количество пропущенных значений:\n{}\n".format(evaluated_dataset.isnull().sum().sum()) + '_' * 50+ '\n')
    
    evaluated_dataset.isnull().sum().plot(kind='bar')
    # Add a title and show the plot.
    plt.title('Распределение пропущенных значений по столбцам')
    # Create tick mark labels on the Y axis and rotate them.
    plt.xticks(rotation = 90)
    # Create X axis label.
    plt.xlabel("Columns")
    # Create Y axis label.
    plt.ylabel("NaN Values");
    
    # Количество значений, отличных от NaN
    print("Количество значений, отличных от NaN:\n{}\n".format(evaluated_dataset.count().sum()) + '_' * 50+ '\n')
    
    # Поиск дублирующих строк 
    print("Дублирующих строк:\n{}\n".format(evaluated_dataset.duplicated().sum()) + '_' * 50+ '\n')
    
    # Форма нарбора данных
    print("Форма нарбора данных:\n{}\n".format(evaluated_dataset.shape) + '_' * 50+ '\n')
   
    # Тип индекса набора данных
    print("Тип индекса набора данных:\n{}\n".format(evaluated_dataset.index) + '_' * 50+ '\n')
````

````python
verify_data_quality(sber_market_db)
````

````python
# Удаление дублирующих значений
sber_market_db = sber_market_db.drop_duplicates()

# Проверка результата
print("Дублирующих строк: {}".format(sber_market_db.duplicated().sum()))
````

````python
sber_market_db['state_replacements'] = sber_market_db['state_replacements'].cat.add_categories('completed')
sber_market_db['state_cancellations'] = sber_market_db['state_cancellations'].cat.add_categories('completed')
````

````python
sber_market_db['state_replacements'] = sber_market_db['state_replacements'].fillna('completed')
sber_market_db['state_cancellations'] = sber_market_db['state_cancellations'].fillna('completed')

sber_market_db['item_id_replacements'] = sber_market_db['item_id_replacements'].fillna(0)
sber_market_db['item_id_cancellations'] = sber_market_db['item_id_cancellations'].fillna(0)
````

````python
verify_data_quality(sber_market_db)
````

````python
sber_market_db = sber_market_db.dropna()
````

````python
verify_data_quality(sber_market_db)
````
